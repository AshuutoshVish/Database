```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:Jobma009%40@localhost:3306/jobma")
```

# Extracting Data from database

```python
#For clint details
jobma_catcher_df = pd.read_sql("SELECT * FROM jobma_catcher", engine) #Client details

jobma_login_df = pd.read_sql("SELECT * FROM jobma_login", engine) #Usage details
jobma_wallet_df = pd.read_sql("SELECT * FROM wallet", engine) #How aggresively using + how much #account Wallet info
jobma_subscription_df = pd.read_sql("SELECT * FROM subscription_history", engine) #how frequently using #subscription history
print("Done start")


jobma_catcher_language_df = pd.read_sql("SELECT * FROM jobma_catcher_language", engine) #Account setttings 
ai_video_kit_df = pd.read_sql("SELECT * FROM ai_video_kit", engine) #Features using setttings (How much) #Kits created by employers for AI Live interview
jobma_pitcher_applied_jobs_df = pd.read_sql("SELECT * FROM jobma_pitcher_applied_jobs", engine) #How many jobs he is posting #Pitcher/Candidate details and status.
jobma_pitcher_df = pd.read_sql("SELECT * FROM jobma_pitcher", engine) #How many candidate #Pitcher/Candidate profile details.
print("Done")

jobma_employer_job_posting_df = pd.read_sql("SELECT * FROM jobma_employer_job_posting", engine) #how many positions he has created
jobma_pitcher_invitations_df = pd.read_sql("SELECT * FROM jobma_pitcher_invitations", engine) #Interview 
print("Done")

jobma_interviews_df = pd.read_sql("SELECT * FROM jobma_interviews", engine) #candidate interview response data (just for pre-recorded type)
jobma_interviews_online_df = pd.read_sql("SELECT * FROM jobma_interviews_online", engine) ##candidate interview response data (just for live interview type)
jobma_ai_interviews_df = pd.read_sql("SELECT * FROM jobma_ai_interviews", engine) #candidate interview response data
print("Prefinal Done")


report_df = pd.read_sql("SELECT * FROM report", engine) #different types of saved reports.
shared_report_df = pd.read_sql("SELECT * FROM shared_report", engine) #subscription history #Share with sub-account or with main account.
print("Final Done")
```
